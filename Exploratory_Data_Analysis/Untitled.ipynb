# Exploratory Data Analysis (EDA)
Exploratory Data Analysis (EDA) is the practice of examining datasets to summarize their main characteristics, often using visual methods. It involves both graphical and non-graphical techniques and is used to maximize insight into a dataset, identify any underlying structures, detect outliers or anomalies, and form hypotheses that could be tested in subsequent modeling steps.

In this module, we will focus on the following:

- Summary statistics using Pandas,
- Visualizing distributions,
- Identifying outliers and anomalies,
- Understanding correlations and relationships,
- Working with categorical variables.

```python
# summary Statistics Using pandas 
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
data = {
    "ProducID": [101, 102, 103, 104],
    "SalesAmount": [101, 102, 103, 104],
    "Region": ["North", "South", "West", "East"]
}
```

```python
df = pd.DataFrame(data)
```

```python
# summary statistics for numerical data 
summary = df.describe()
print(summary)
```

```python
print(df['SalesAmount'].mean())
```

## Visualizing Data Distributions

Visualizing data is essential for understanding its underlying structure, trends, and patterns. Graphical representations such as histograms, box plots, and scatter plots are useful for summarizing and analyzing data distributions. We will cover data visualization and its essential libraries in more detail in the next course, where we'll explore various visual charts. For now, this section will briefly introduce the basic concepts and code to help you get familiar with them.

##### Histograms

A histogram is one of the simplest and most effective ways to understand the distribution of a numerical variable. It shows how the values of a variable are distributed across different bins (or ranges).

```python
data = {'SalesAmount': [200, 150, 400, 300, 250, 350, 450, 500]}
df = pd.DataFrame(data)
```

```python
# ploting a histogram of slaesamount

plt.hist(df["SalesAmount"], bins=5, color='blue', edgecolor='black')
plt.title('slaes amount distribution')
plt.xlabel('slaes amount')
plt.ylabel('Frequency')
plt.show()
```

## Box Plots

A box plot (or whisker plot) is used to visualize the spread of the data and identify outliers. It shows the minimum, first quartile (25th percentile), median, third quartile (75th percentile), and maximum values of the dataset.

```python
# Box plot for Salesamount

sns.boxplot(x=df['SalesAmount'])
plt.title('Box plot of sales Amount')
plt.show()
```

## Scatter Plots

A scatter plot shows the relationship between two numerical variables. Each point in the plot represents an observation in the dataset, with the x and y axes representing two variables.

```python
# Scatter plot of SalesAmount vs another variable (e.g., ProductID)
df['ProductID'] = [101, 102, 103, 104, 105, 106, 107, 108]
plt.scatter(df['ProductID'], df['SalesAmount'], color='green')
plt.title("product vs amount")
plt.xlabel('ProductID')
plt.ylabel('Sales Amount')
plt.show()
```

## Pair Plots

Pair plots provide a way to visualize pairwise relationships between multiple variables in the dataset. It plots every numerical variable against every other numerical variable.

```python
# Pair plot of the numerical columns in the dataset
sns.pairplot(df)
plt.show()
```

## Heatmaps for Correlation

A heatmap is a visual representation of a correlation matrix that shows the strength of relationships between numerical variables. This is particularly useful when you want to understand the correlation between multiple variables.

```python
# corellation matrics 
corr_matrix = df.corr()
# Heatmap of the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('correlation heatmap')
plt.show()
```
